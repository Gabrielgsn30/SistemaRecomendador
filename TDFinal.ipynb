```python
#utilizado para calcular distancias
import math
raioDaTerra   =  6371 
```

```python
def converterGrausParaRad(numero):
    rad = (numero/180)*math.pi
    return rad
```

```python
def calcularDistancia (lat_ref,lon_ref, lat_posto, lon_posto):
  distancia   = raioDaTerra * math.acos( 
      math.cos( converterGrausParaRad(lat_ref) )
    * math.cos( converterGrausParaRad(lat_posto) ) 
    * math.cos( converterGrausParaRad(lon_ref ) - converterGrausParaRad(lon_posto) )
    + math.sin( converterGrausParaRad(lat_ref) ) 
    * math.sin( converterGrausParaRad(lat_posto) )
  )
  return distancia
```

```python
""" 
Em Python, arrays são chamados de Listas.
Podemos usar uma função do Python para ordenar os valores de distância em ordem
cresente. Desta forma, a função menor irá funcionar independente da quantidade
de distâncias que passamos.
"""
def menor(listaDePostos):
  # print("LISTA DE POSTOS",listaDePostos)
  auxiliar    = listaDePostos[:]
  auxiliar.sort()               #Ordena de forma crescente
  menorValor  = auxiliar[0]
  #return [(n, listaDePostos[n]) for n, x in enumerate(listaDePostos) if x==menorValor]
  indiceValor = listaDePostos.index(menorValor)
  idByIndice  = indiceValor + 1
  #print("LISTA AUXILIAR", auxiliar)
  return (idByIndice)
          #, menorValor)
```

```python
arq_csv["afinidade"] = 'NaN'
arq_csv.sample(6)
```

Filtrando a tabela por carros para obter o valor de afinidade.

```python
for carro in ids_carros:
  filtro = arq_csv['veiculos_eletricos_id'] == carro
  posições = np.flatnonzero(filtro)
  tabela_filtrada = arq_csv.iloc[posições]

  tamanho_total = tabela_filtrada.shape[0] # Total de registros de um carro

  """
    A partir deste ponto tenho uma tabela para cada carro.
    Agora posso filtrar em cada tabela de carro os postos de
    gasolina abastecidos por ele e calcular sua afinidade.
  """

  ids_postos = tabela_filtrada['idposto']
  ids_postos = ids_postos.drop_duplicates()
  ids_postos = ids_postos.tolist()            # ID's de postos abastecidas por um carro

  """
    Para cada posto, irei criar uma tabela filtrada para obter a quantidade de
    vezes que um carro abasteceu neste posto.
  """
  
  for posto in ids_postos:
    filtro = tabela_filtrada['idposto'] == posto
    posições = np.flatnonzero(filtro)
    tabela_filtrada_por_posto = tabela_filtrada.iloc[posições]

    ocorrencias = tabela_filtrada_por_posto.shape[0] # Ocorrências de abastecimento em determinado posto

    print("O carro {} abasteceu no posto {}, um total de {} veze(s)"
      .format(carro,posto,ocorrencias),end='')

    """
      A partir deste ponto posso calcular relação de afinidade 
      de cada carro com um posto abastecido
    """
    
    afinidade = ocorrencias/float(tamanho_total) 
    print('[ {}% ]'.format(int(afinidade*100)))

    """
      Agora basta adicionar ao csv, o valor de afinidade encontrado.
    """

    arq_csv.loc[arq_csv.idposto == posto,'afinidade'] = afinidade
    arq_csv['afinidade'] = arq_csv['afinidade'].astype(float)
```

```python
class MF(nn.Module):
    def __init__(self, num_users, num_items,emb_size=100000):
        super(MF, self).__init__()
        self.user_emb = nn.Embedding(emb_size ,num_users )
        self.item_emb = nn.Embedding(emb_size ,num_users )
        print(self.user_emb)
        print(self.item_emb)
        # initializing our matrices with a positive number generally will yield better results
        self.user_emb.weight.data.uniform_(0, 0.5)
        self.item_emb.weight.data.uniform_(0, 0.5)
        
    def forward(self, u, v):
        u = self.user_emb(u)
        v = self.item_emb(v)
        return (u*v).sum(1)  # taking the dot product
```

```python
users_ratings = arq_csv.copy()
#game_ratings = game_ratings.drop(columns=['id', 'Latitude','Longitude','Charge','abastecido','afinidade'], axis=1)
users_ratings.reset_index(drop=True)
```

```python
num_users = len(users_ratings['veiculos_eletricos_id'].unique())
num_stations = len(users_ratings['idposto'].unique())
model = MF(num_users, num_stations, emb_size=100000)
```

```python
train_df, valid_df = train_test_split(users_ratings, test_size=0.4)

# resetting indices to avoid indexing errors
train_df = train_df.reset_index(drop=True)
test_df = valid_df.reset_index(drop=True)
#print(train_df.max())
```

```python
def train_epocs(model, epochs=10, lr=0.01, wd=0.0):
    optimizer = torch.optim.Adam(model.parameters(), lr=lr, weight_decay=wd)
    model.train()
    for i in range(epochs):

        usernames = torch.LongTensor(train_df.veiculos_eletricos_id.values)
        id_station = torch.LongTensor(train_df.idposto.values)
        ratings = torch.FloatTensor(train_df.avaliacao.values)
        y_hat = model(usernames, id_station)
        loss = F.mse_loss(y_hat, ratings)
        optimizer.zero_grad()  # reset gradient
        loss.backward()
        optimizer.step()
        print(loss.item())
    test(model)
```

```python
def test(model):
    model.eval()
    user = torch.LongTensor(test_df.veiculos_eletricos_id.values)
    id_station = torch.LongTensor(test_df.idposto.values)
    ratings = torch.FloatTensor(test_df.avaliacao.values)
    y_hat = model(user, id_station)
    loss = F.mse_loss(y_hat, ratings)
    print("test loss %.3f " % loss.item())
```
